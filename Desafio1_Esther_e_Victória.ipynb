# Exercício 2

2. Selecione somente os intervalos contínuos de 1 minuto da coluna Time. Exemplo: o script   deve   selecionar   somente   os   intervalos   contínuos   que   se   iniciem   em HH:mm:00 e terminem em HH:mm:59, onde HH (hora) mm (minuto) devem ser os mesmos no intervalo. 
<br>
Dica: descarte os dados não contínuos do DataFrame.

Convertendo o tipo de MeaValue para float para podermos calcular a média.

```python
df['MeaValue'] = df['MeaValue'].str.replace(',','.').astype(float)
```

Aplicamos o groupby e o grouper para visualizarmos e selecionarmos os intervalos contínuos.

```python
# Aqui vemos que o primeiro intervalo que começa com hh:mm:00 é o 06:41:0
df.groupby(pd.Grouper(key='date_time', freq='1min'))['date_time'].count()
```

E agora que sabemos o intervalo podemos criar um dataframe com os mesmos e seus respectivos segundos.

```python
segundos = pd.DataFrame(list(zip(df.groupby(pd.Grouper(key='date_time', freq='1min'))['date_time'].count().index, 
                                 df.groupby(pd.Grouper(key='date_time', freq='1min'))['date_time'].count())), 
                        columns=['date_time','Segundos'])
segundos.head()
```

Aplicamos a formatação dt.strftime('%Y-%m-%d %H:%M') para igualar as colunas date_time dos dataframes df e segundos, para unificá-los com o merge posteriormente.

```python
df['date_time'] = df['date_time'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df.head()
```

```python
segundos['date_time'] = segundos['date_time'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
segundos.head()
```

```python
df = df.merge(segundos, left_on='date_time', right_on='date_time')
df.head()
```

E agora conseguimos filtrar os intervalos contínuos com o .loc .

```python
df = df.loc[df['Segundos'] == 60]
```

```python
del df['Segundos']
```

```python
df.head()
```

# Exercício 3

3. Em  cada  intervalo   selecionado,  faça   a  média  simples   dos  dados  da   coluna MeaValue.

Aqui calculamos a média de acordo com o grupo do mesmo intervalo.

```python
df.groupby(df['date_time'])['MeaValue'].agg('mean')
```

Criamos um dataframe com os dados acima.

```python
media = pd.DataFrame(list(zip(df.groupby(df['date_time'])['MeaValue'].agg('mean').index, 
                              df.groupby(df['date_time'])['MeaValue'].agg('mean'))), 
                     columns=['date_time','ValorMedio'])
media.head()
```

Conforme as colunas date_time, unimos os dataframes media e df e, por fim, associamos cada média calculada com os integrantes dos respectivos intervalos contínuos.

```python
df = df.merge(media, left_on='date_time', right_on='date_time')
```

```python
df.head()
```

Criamos uma coluna chamada Hora que já recebe o dado formatado para hora e minuto.

```python
df['Hora'] = pd.to_datetime(df['date_time'])
```

```python
df['Hora'] = df['Hora'].dt.strftime('%H:%M')
```

```python
df.head()
```

Renomeamos a coluna Date para Data.

```python
df = df.rename(columns = {'Date':'Data'})
df.head()
```
